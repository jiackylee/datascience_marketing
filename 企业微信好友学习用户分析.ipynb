## 筛选非零数据

### Subtask:
对于每个指定的年份列（'2020_days', '2021_days', '2022_days', '2023_days', '2024_days', '2025_days'），筛选出该列中值大于 0 的行，创建新的 DataFrame 或 Series。

Iterate through the selected columns and filter the DataFrame for non-zero values in each column, storing the results in a dictionary.



```python
filtered_data = {}
for column in selected_columns:
    filtered_data[column] = df[df[column] > 0]

# You can optionally display the head of one of the filtered dataframes to verify
# display(filtered_data['2020_days'].head())
```

## 描述性统计分析

### Subtask:

## 可视化非零数据

# Task
使用 pandas 库读取文件 "/content/企业微信好友数据分析.xlsx"，查看数据的基本信息，对数据进行清洗（处理缺失值、重复值等）。然后，对 '2020_days', '2021_days', '2022_days', '2023_days', '2024_days', '2025_days' 这几个列中非零的数据进行单独分析。最后，按照活跃天数分组（0、1-3、4-10、11-20、21-40、41-100、100+），对这几个列的数据进行分组分析。

## 定义分组区间和标签

### Subtask:
明确活跃天数的分组区间和对应的标签。


**Reasoning**:
Define the bins and labels for grouping active days as requested.



```python
bins = [0, 1, 4, 11, 21, 41, 101, float('inf')]
labels = ['0', '1-3', '4-10', '11-20', '21-40', '41-100', '100+']
```

## 创建新的分类列

### Subtask:
对于每个指定的年份列，根据定义的分组区间，使用 `pd.cut()` 或 `pd.cut()` 创建新的分类列，表示活跃天数所属的分组。


**Reasoning**:
Iterate through the selected columns and create new categorical columns based on the defined bins and labels.



```python
for column in selected_columns:
    df[f'{column}_group'] = pd.cut(df[column], bins=bins, labels=labels, right=False)

# Display the first few rows with the new group columns
display(df[['user_Id'] + [f'{col}_group' for col in selected_columns]].head())
```

## 按年份和分组进行计数

### Subtask:
对于每个年份，按新创建的分类列进行分组，并计算每个分组中的用户数量。


**Reasoning**:
Iterate through the selected columns, group the DataFrame by the corresponding group column, count the number of users in each group, and store the counts in a dictionary.



```python
grouped_data = {}
for column in selected_columns:
    group_column = f'{column}_group'
    grouped_data[column] = df.groupby(group_column)['user_Id'].count()

# Print or display the counts for each year's groups
for year_column, group_counts in grouped_data.items():
    print(f"\n用户数量按活跃天数分组 ({year_column}):")
    print(group_counts)
```

## 可视化分组结果

### Subtask:
绘制柱状图或堆积柱状图，展示每个年份不同活跃天数分组的用户分布情况。


**Reasoning**:
Create bar plots for each year's active day group distribution.



```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

for year_column, group_counts in grouped_data.items():
    year = year_column.split('_')[0]
    plt.figure(figsize=(10, 6))
    group_counts.plot(kind='bar')
    plt.title(f'User Distribution by Active Days Group in {year}')
    plt.xlabel('Active Days Group')
    plt.ylabel('Number of Users')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

## Summary:

### Data Analysis Key Findings

*   The data was successfully grouped by active day ranges (0, 1-3, 4-10, 11-20, 21-40, 41-100, 100+) for each year from 2020 to 2025.
*   The analysis provided a count of users within each active day group for every specified year.
*   Bar charts were generated for each year, visually representing the distribution of users across the different active day groups.

### Insights or Next Steps

*   Analyze the trend of user distribution across active day groups over the years to identify changes in overall user engagement.
*   Investigate the characteristics of users in the "0" active days group for each year to understand reasons for inactivity and potential re-engagement strategies.


## 横向对比各年份分组数据趋势

### Subtask:
将各年份按活跃天数分组的用户数量数据进行整合，以便进行趋势分析和可视化。

**Reasoning**:
Combine the grouped data for each year into a single DataFrame to facilitate comparison of user distribution across active day groups over time.

```python
# Combine the grouped data into a single DataFrame for easier plotting
grouped_df = pd.DataFrame(grouped_data)

# Reset index to make the active days group a column
grouped_df = grouped_df.reset_index()

# Rename the index column for clarity
grouped_df = grouped_df.rename(columns={'index': 'Active_Days_Group'})

# Display the combined data
display(grouped_df)
```

## 可视化分组数据趋势

### Subtask:
绘制图表（如折线图或堆积柱状图）展示各年份不同活跃天数分组的用户数量趋势。

**Reasoning**:
Plot the combined grouped data to visualize the trends of user distribution across different active day groups from 2020 to 2025. A melted DataFrame is needed for plotting with seaborn.

```python
# Melt the DataFrame to long format for easier plotting with seaborn
grouped_df_melted = grouped_df.melt(id_vars='Active_Days_Group', var_name='Year', value_name='Number_of_Users')

# Extract the year from the column names
grouped_df_melted['Year'] = grouped_df_melted['Year'].str.replace('_days', '')

# Plot the trend of user distribution across active days groups over the years
plt.figure(figsize=(12, 7))
sns.lineplot(data=grouped_df_melted, x='Year', y='Number_of_Users', hue='Active_Days_Group', marker='o')
plt.title('Trend of User Distribution by Active Days Group (2020-2025)')
plt.xlabel('Year')
plt.ylabel('Number of Users')
plt.grid(True)
plt.show()
```

## Summary:

### Data Analysis Key Findings

* The user distribution across different active day groups has been successfully visualized over the years (2020-2025).

### Insights or Next Steps

* Analyze the line plot to identify which active day groups have grown or shrunk over the years.
* Investigate factors that might have contributed to the observed trends in user activity.
